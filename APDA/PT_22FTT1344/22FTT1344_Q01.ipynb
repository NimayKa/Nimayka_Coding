Question 1 A

```python
import pandas as pd
import numpy as np


df1 = pd.read_csv('student_pt1.csv')
df2 = pd.read_csv('student_pt2.csv')
df3 = pd.read_csv('student_pt3.csv')

print(df1)
print('\n\n',df2)
print('\n\n',df3)
print(df1.columns)
print(df2.columns)
print(df3.columns)
```

Question 1 B I)

```python
student_combined = df1.merge(df2, how='outer')
print(student_combined)
```

Question 1 B II)

```python
studentRows = student_combined.merge(df3, how='outer',on='id')
print (studentRows)
print (studentRows.columns)
print (studentRows.info())
```

Question 1 C)

```python
studentRows['gender'] = studentRows['gender'].str.replace('0', 'NaN').str.replace('none', 'NaN')
studentRows['race/ethnicity'] = studentRows['race/ethnicity'].str.replace('0', 'NaN').str.replace('none', 'NaN')
studentRows['parental level of education'] = studentRows['parental level of education'].str.replace('0', 'NaN').str.replace('none', 'NaN')
studentRows['lunch'] = studentRows['lunch'].str.replace('0', 'NaN').str.replace('none', 'NaN')
studentRows['test preparation course']=studentRows['test preparation course'].str.replace('none', 'NaN')
studentRows['math score'] = studentRows['math score'].replace(0, np.nan)
studentRows['reading score'] = studentRows['reading score'].replace(0, np.nan )
studentRows['writing score'] = studentRows['writing score'].replace(0, np.nan)
print (studentRows)
```

Question 1 D)

```python
total_missing = studentRows.isnull().sum().sum()
for i in studentRows.columns:
    missing_count = studentRows[i].isnull().sum()
    if missing_count >= 1:
        print(f"Missing Value in {i} = {missing_count}")

print(f"Total missing values = {total_missing}")
```

```python
studentRows = studentRows.drop(columns=['test preparation course'])
```

Question 1 E)

```python
studentRows.drop(studentRows[studentRows.isna().sum(axis=1) >= 3].index, inplace=True)
print (studentRows)
```

```python
print('\nDisplaying all Missing Data\n')
df_missingvalue_1row = studentRows[studentRows.isna().sum(axis=1)>=1]
df_missingvalue_1row
```

Question 1 F)

```python
studentRows.dropna(subset=['math score'], inplace=True)
student_grade = ['E','D','C','B','A']
studentRows['student_grade'] = pd.cut(studentRows['math score'], [0,20,40,60,80,100], labels=student_grade)
```
