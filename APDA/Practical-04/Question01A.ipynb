```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Iris.csv",index_col=['Id'])
df
```

Histrogram

```python
attribute = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
for i in attribute:
    plt.figure(figsize=(8, 6))
    plt.hist(df[i], bins=10, edgecolor='black') 
    plt.title(f'Histogram of {i}')
    plt.xlabel(i)
    plt.ylabel('Frequency')
    plt.show()
```

PairPlot

```python
sns.pairplot(df[attribute])
plt.show()
```

Scatter Plot SepalLength vs PetalLength

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['SepalLengthCm'], df['PetalLengthCm'], color='blue', alpha=0.2)
plt.title('Scatter Plot: Sepal Length vs. Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.grid(True)
plt.show()
```

Scatter Plot SepalWidth vs PetalWidth

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['SepalWidthCm'], df['PetalWidthCm'], color='blue', alpha=0.2)
plt.title('Scatter Plot: Sepal Width vs. Petal width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.grid(True)
plt.show()
```
