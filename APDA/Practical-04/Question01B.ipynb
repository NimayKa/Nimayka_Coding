```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("pokemon_data.csv", index_col = ['#'])
df
```

Exract Pokemon who has fire type only | applying data visualization on fire and water type | Histogram


Attack Attribute:

Fire Type: Fire type Pokémon have a higher mean value in the Attack attribute compared to Water type Pokémon. Fire type Pokémon tend to have stronger offensive capabilities in terms of attack power.


Speed Attribute:

Fire Type: Fire type Pokémon also have a higher mean value in the Speed attribute compared to Water type Pokémon. Fire type Pokémon are faster and may have an advantage in terms of acting earlier in battles or escaping from opponents.
HP Attribute:

Water Type: Water type Pokémon have a higher mean value in the HP (Hit Points) attribute compared to Fire type Pokémon.Water type Pokémon tend to have higher health pools, making them potentially more resilient and able to withstand attacks for longer.

Defense Attribute:

Water Type: Water type Pokémon also have a higher mean value in the Defense attribute compared to Fire type Pokémon. Water type Pokémon are better at withstanding physical attacks and have stronger defensive capabilities.

```python
filtered_df = df[(df["Type 1"].isin(["Fire", "Water"]))]

# Group the data by Type 1 and calculate the mean for selected attributes
grouped_df = filtered_df.groupby("Type 1")[['HP', 'Attack', 'Defense', 'Speed']].mean()

# Plotting the bar chart
grouped_df.plot(kind='bar', figsize=(10, 6))
plt.title('Mean Attribute Values for Fire and Water Type Pokémon')
plt.xlabel('Type')
plt.ylabel('Mean Value')
plt.xticks(rotation=0)
plt.legend(title='Attributes')
plt.grid(axis='y')
plt.show()
```

Water Type, Generation 3: 

In the scatter plot, "Water" type Pokémon in Generation 3 tend to have higher HP values compared to other generations.Water type Pokémon Gen 3 were designed with greater health pools, potentially making them more resilient in battles.

Fire Type, Generation 6:

On the other hand, among "Fire" type Pokémon, those from Generation 6 appear to have higher HP values. Fire type Pokémon Gen 6 were designed with increased hit points, contributing to their overall durability and ability to withstand attacks.

```python
attribute_x = 'HP'
attribute_y = 'Generation'

plt.figure(figsize=(8, 6))
plt.scatter(filtered_df[filtered_df["Type 1"] == "Fire"][attribute_x],filtered_df[filtered_df["Type 1"] == "Fire"][attribute_y],color='r', label='Fire', alpha=0.7)
plt.scatter(filtered_df[filtered_df["Type 1"] == "Water"][attribute_x],filtered_df[filtered_df["Type 1"] == "Water"][attribute_y],color='b', label='Water', alpha=0.7)
plt.title(f'Scatter Plot: {attribute_x} vs {attribute_y}')
plt.xlabel(attribute_x)
plt.ylabel(attribute_y)
plt.legend()
plt.grid(True)
plt.show()
```
