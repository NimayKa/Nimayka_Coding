```python
# Import necessary libraries
import numpy as np
import pandas as pd

# To display full data of each columns
pd.set_option('display.max_columns', None)

# To bypass a warning regarding mixed data types
AB = pd.read_csv('Airbnb_Open_Data.csv',low_memory=False)
AB.head(5)
```

```python
# To correct the mistake in 'neighbourhood group'

AB.loc[AB['neighbourhood group']=='manhatan', 'neighbourhood group'] = 'Manhattan'
AB.loc[AB['neighbourhood group']=='brookln', 'neighbourhood group'] = 'Brooklyn'
AB['neighbourhood group'].unique()
```

```python
# To remove unwanted symbols in 'price' and 'service fee'

AB['price'] = AB['price'].str.replace('$', '', regex = True).str.replace(',', '', regex = True)
AB['service fee'] = AB['service fee'].str.replace('$', '', regex = True)
AB.head(5)
```

```python
# To change the data type of 'price' and 'service fee'

AB['price'] = AB['price'].astype('float64')
AB['service fee'] = AB['service fee'].astype('float64')
```

```python
# To display the data type of each column

AB.dtypes
```

```python
# To find out the number of missing values of each column

AB.isnull().sum()
```

```python
# Drop unecessary columns in the dataset

AB.drop('license', axis='columns', inplace=True)
AB.drop('house_rules', axis='columns', inplace=True)
AB.drop('country code', axis='columns', inplace=True)
AB.drop('id', axis='columns', inplace=True)
AB.drop('last review', axis='columns', inplace=True)
AB.drop('host id', axis='columns', inplace = True)
AB.drop('host name', axis='columns', inplace = True)
AB.drop('country', axis='columns', inplace = True)
```

```python
# Drop rows that contain missing values in some columns

AB.dropna(subset=['neighbourhood','instant_bookable','cancellation_policy','Construction year','availability 365'], inplace=True)
```

```python
# To replace the missing values in 'host_identity_verified' to become 'Not Stated'

AB['host_identity_verified'].fillna('Not Stated', inplace = True)
AB['NAME'].fillna('Unknown', inplace = True)
AB['number of reviews'].fillna(1.0, inplace = True)
AB['calculated host listings count'].fillna(1.0, inplace = True)

AB.head(5)
```

```python
# To replace missing values with specific values

x = AB['reviews per month'].mean()
y = AB['review rate number'].mean()

AB['reviews per month'].fillna(x, inplace = True)
AB['review rate number'].fillna(y, inplace = True)
AB['minimum nights'].fillna(1.0, inplace = True)

AB.head(5)
```

```python
#Filter Missing Value Out
ABFilter = AB[AB.notnull()]
```

```python
#Filling missing value for Price and Service Fees
ABFilter = AB[AB.notnull()]
moderoom_type_price = ABFilter.groupby(['room type'])['price'].apply(lambda x: x.mode().iloc[0])
moderoom_type_servicefee = ABFilter.groupby(['room type'])['service fee'].apply(lambda x: x.mode().iloc[0])

AB['price'].fillna(AB['room type'].map(moderoom_type_price), inplace=True)
AB['service fee'].fillna(AB['room type'].map(moderoom_type_servicefee), inplace=True)
```

```python
ABFilter = AB[AB.notnull()]
AB[['lat', 'long']] = AB[['lat', 'long']].round(2)

#Filling missing value for Latitude and Longtitude
mode_neighbourhood_lat = ABFilter.groupby(['neighbourhood'])['lat'].apply(lambda x: x.mode().iloc[0])
mode_neighbourhood_long = ABFilter.groupby(['neighbourhood'])['long'].apply(lambda x: x.mode().iloc[0])

AB['lat'].fillna(AB['neighbourhood'].map(mode_neighbourhood_lat), inplace=True)
AB['long'].fillna(AB['neighbourhood'].map(mode_neighbourhood_long), inplace=True)
```

```python
ABFilter = AB[AB['neighbourhood group'].notnull()]
neighbourhood_group = ABFilter.groupby('neighbourhood')['neighbourhood group'].apply(lambda x: x.iloc[0])
AB['neighbourhood group'].fillna(AB['neighbourhood'].map(neighbourhood_group), inplace=True)
```

```python
AB = AB.rename(columns = ({'availability 365':'availability',
                           'host_identity_verified':'verification',
                           'instant_bookable':'bookable',
                           'cancellation_policy':'cancellation',
                           'review rate number':'rate',
                           'calculated host listings count': 'total host listings',
                           'NAME':'airbnb',
                           'neighbourhood group':'district'}))

AB.head()
```

```python
# Create new dataset after data cleaning

AB.to_csv('airbnb_new.csv', index=False)
```
