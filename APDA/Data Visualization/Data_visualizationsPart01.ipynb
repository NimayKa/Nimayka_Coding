```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

tips = pd.read_csv("tips.csv")

#create a cross tabulation by day and party size
party_counts = pd.crosstab(tips['day'],tips['size'])

#select party sizes from 2 to 5 
party_counts = party_counts.loc[:,2:5]

print(party_counts)

# Normalize the data so that the count of each party size per row sums to 1 for each day
row_sums = party_counts.sum(axis=1)

print(row_sums)

# div = divide party_counts with row_sums to get percentage value
party_pcts = party_counts.div(row_sums, axis=0) 
# change it to percentage
party_pcts_percent = party_pcts * 100

print(party_pcts_percent)

# Create a stacked bar plot of the normalized data
party_pcts_percent.plot.bar()

# Set plot labels and title
plt.xlabel('Day')
plt.ylabel('Percentage')
plt.title('Percentage of Party Size by Day')

# Display theÂ plot
plt.show()

```

```python

#load the tips dataset using read-csv
tips = pd.read_csv('tips.csv')

# call column and calculate the tip percentage
tips ['tips_pct'] = tips['tip'] / (tips ['total_bill'] - tips ['tip'])

# set the plotting style
sns.set(style='darkgrid')
# create a horizontal bar plot using seaborn
sns.barplot(x='tips_pct',y='day',data=tips,orient='h')

# Set plot label anbd title
plt.xlabel('Tip percentage')
plt.ylabel('Day')
plt.title('Tip percentage by day')
plt.show()
```
