```python
import pandas as pd
import numpy as np

# Create a sample DataFrame
df = pd.read_csv("hyundi.csv")
pd.set_option('display.max_rows', 10)

# Randomly select 10 indices
random_indices = np.random.choice(df.index, size=3)
print("Random Number:",random_indices)

# Introduce missing values at the selected indices
for index in random_indices:
    col = np.random.choice(df.columns)
    df.at[index, col] = np.nan

# Print the modified DataFrame with missing values
print(df)

Total = 0
for i in df.columns:
    if df[i].isnull().sum() >= 1:
        print("Missing Value in",i,"=", df[i].isnull().sum())
        Total += df[i].isnull().sum()
print("Total missing values =", Total)

print('\nGrabbing NA Data\n')
df_missingvalue = df[df.isna().any(axis=1)]
print(df_missingvalue)


df.iloc[random_indices] = df_missingvalue

#df.to_csv('newdataset.csv', index = False)

```

```python
import pandas as pd
import numpy as np

# Create a sample DataFrame
df = pd.read_csv("hyundi.csv")

# Introduce missing values with a chance of 1 or 2 missing values in each row
num_missing_rows = np.random.choice(df.index, size=20, replace=False)  # Select 10 random rows

for row_index in num_missing_rows:
    num_missing = np.random.choice([1, 2])  # Randomly choose 1 or 2 missing values
    missing_columns = np.random.choice(df.columns, num_missing, replace=False)  # Select random columns

    for col in missing_columns:
        df.at[row_index, col] = np.nan

# Save the DataFrame with missing values to a CSV file
#df.to_csv("data_with_missing_values.csv", index=False, na_rep='NaN')

# Print the DataFrame
print(df)
Total = 0
for i in df.columns:
    if df[i].isnull().sum() >= 1:
        print("Missing Value in",i,"=", df[i].isnull().sum())
        Total += df[i].isnull().sum()
print("Total missing values =", Total)

print('\nGrabbing Missing Data\n')
df_missingvalue = df[df.isna().any(axis=1)]
print(df_missingvalue)
```
