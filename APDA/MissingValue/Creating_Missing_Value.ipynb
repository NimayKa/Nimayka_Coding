```python
import pandas as pd
import numpy as np

df = pd.read_csv("Adidas US Sales Datasets.csv")

num_missing_rows = np.random.choice(df.index, size=50, replace=False) 

for row_index in num_missing_rows:
    num_missing = np.random.choice([1,2,3]) 
    missing_columns = np.random.choice(df.columns, num_missing, replace=False)

    for col in missing_columns:
        df.at[row_index, col] = np.nan

df.to_csv("data_with_missing_values.csv", index=False, na_rep='NaN')

print(df)
Total = 0
for i in df.columns:
    if df[i].isnull().sum() >= 1:
        print("Missing Value in",i,"=", df[i].isnull().sum())
        Total += df[i].isnull().sum()
print("Total missing values =", Total)



print('\nDisplaying all Missing Data\n')
df_missingvalue = df[df.isna().any(axis=1)]
print(df_missingvalue)

print('\nDisplaying 1 Column Missing Data\n')
rows_with_1_missing_value = df[df.isnull().sum(axis=1) == 1]
print (rows_with_1_missing_value)
```
