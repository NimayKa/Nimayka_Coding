```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x= [1,2,3]
y= [2,4,6]

plt.plot(x,y)
plt.show()
```

```python
x= [1,2,3]
y= [2,4,6]
plt.plot(x,y,label='line hijau',color='green',marker= '*',markersize=10,markeredgecolor='red',linestyle='--')

plt.xlabel('axis for x',loc = 'center', fontdict ={'fontname':'Comic Sans MS','fontsize':20})
plt.ylabel('axis for y')
plt.title('Our First Graph')


plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10,12])

plt.legend()

plt.show()
```

```python
x= [1,2,3,4]
y= [2,4,6,8]


plt.plot(x[:2],y[:2],'b',label= 'Biru')
plt.plot(x[1:],y[1:],'b--')

plt.xlabel('axis for x')
plt.ylabel('axis for y')
plt.title('Our First Graph')


plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10,12])

plt.legend()

plt.show()
```

```python
a = [1,2,3,3.5]
b = [3,7,10,4]

x= [1,2,3]
y=[2,4,6]

plt.figure(figsize=(3,2),dpi=150)
plt.plot(a,b , 'r*--',label= 'Biru')
plt.plot(x,y , label= 'Merah')

plt.title('Our 2nd Graph',loc = 'Center')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10,12])
plt.savefig('Mynewgraph01.png',dpi=200)

plt.legend()

plt.show()
```

```python
labels = ['A','B','C']
values = [1,3,4]

plt.figure(figsize=(6,6))

MyBar = plt.bar(labels,values)
plt.xticks(['A','B','C'])
plt.yticks([1,3,5,7])

MyBar[0].set_hatch('/')
MyBar[2].set_hatch('*')

plt.show()
```

```python
labels = ['A','B','C','D','E']
values = [1,3,4,2,5]

plt.figure(figsize=(6,6))

MyBar = plt.scatter(labels,values,label='blue')

plt.legend()
plt.show()
```
