```python
import pandas as pd

df = pd.read_csv("class-grades.csv")
df.set_index('Prefix',inplace=True)
bin_edges = [0, 20, 40, 60, 80, 100]
bin_labels = ['E', 'D', 'C', 'B', 'A']

df['FinalR'] = pd.cut(df['Final'], bins=bin_edges)
df['Final_R'] = pd.cut(df['Final'], bins=bin_edges, labels=bin_labels)
df.rename(columns={'Final':'Old_final','FinalR':'Final','Final_R':'FinalR'},inplace=True)
df.drop(columns=['Old_final'], inplace=True)
print (df)
```

```python
df.columns
new_labels = {
    'E': 'Fail',
    'D': 'Pass',
    'C': 'Merit',
    'B': 'Merit',
    'A': 'Distinction'
}
finalR_df = df['FinalR']
df['NewLabel']  = finalR_df.map (new_labels)
df
```

```python
df['Assignment'] = df['Assignment'].map(lambda x:x/ 100).round(2)
df
```
