Line Plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#create a numpy array with 10 random numbers and compute the cumulative sum
data = np.random.randn(10).cumsum() #.cumsum() is to create a sequemce of cumulative value

#create a pandas series using the computed data
index = np.arange(0,100,10) #generates array from 0-90 with step of 10
s = pd.Series(data,index=index)

print ('\n', s)

#plot the series as line plot
s.plot()# by default, x-axis index and y-axis is data

# cusomtize the plot 
plt.title("Cumulative Sum of Random Data")
plt.xlabel("Index")

plt.ylabel("Cumulative Sum")
plt.grid(True)

# display the plot
plt.show()
```

Line plot with multiple line

```python
#create a numpy two dimensional array of random numbers
data = np.random.randn(10,4).cumsum(0)# compute cumulative sum in index(0)
#Create a dataframe with cumulative sum data , specifying column names and an index
columns = ['A','B','C','D']
index =  np.arange(0,100,10)
df = pd.DataFrame(data,columns=columns,index=index)
#use the plot)_ method pof datadrame to create a line plot wioth multiple line
df.plot()
```

Subplot

```python
data = np.random.randn(10,4).cumsum(0)
columns = ['A','B','C','D']
index = np.arange(0,100,10)

df = pd.DataFrame(data, columns=columns,index=index)

fig, axes =plt.subplots(nrows=2,ncols=2,figsize=(10,6))

df['A'].plot(ax=axes[0,0],title='Column A')
df['B'].plot(ax=axes[0,1],title='Column B')
df['C'].plot(ax=axes[1,0],title='Column C')
df['D'].plot(ax=axes[1,1],title='Column D')

plt.tight_layout()

plt.show()
```

Bar and Barh Plot

```python
#creates a matplotlib figure containing a 2x1 grid of sublplots
fig, axes = plt.subplots(2,1)#there will be two subplots stacked vertically

#create a pandas series with random data and acustom index
data = pd.Series(np.random.rand(16),index=list('acbdefghijklmnop'))

#plot a vertical bar plot on the first subplot (axes[0])
data.plot.bar(ax=axes[0],color='k',alpha=0.7)
#plot a vertical bar plot on the first subplot (axes[1])
data.plot.barh(ax=axes[1],color='k',alpha=0.7)
#display the plots
plt.show()
```
