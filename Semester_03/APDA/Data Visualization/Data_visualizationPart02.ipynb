Scatter or point plots

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels as sm

macro = pd.read_csv('macrodata.csv')
data = macro[['cpi','m1','tbilrate','unemp']]
trans_data = np.log(data).diff().dropna()
trans_data[-5:]

sns.regplot(x= 'm1',y ='unemp',data=trans_data)
plt.title('Changes in log %s versus log %s'%('m1','unemp'))

sns.pairplot(trans_data,diag_kind='kde',plot_kws={'alpha':0.1})
plt.show()
```

FaceT Grid and Catergorical Data

```python
#load the tips dataset using read-csv
tips = pd.read_csv('tips.csv')

# call column and calculate the tip percentage
tips ['tips_pct'] = tips['tip'] / (tips ['total_bill'] - tips ['tip'])

sns.catplot(x='day', y='tips_pct', hue='time', col='smoker', kind='bar', data=tips[tips.tips_pct < 1])

# set the plotting style
sns.set(style='darkgrid')

# Set plot label anbd title
plt.title('Faceted Bar Plot of Tip percentage by Day, Time, and Smoker')
plt.show()  
```

```python

sns.catplot(x='day', y='tips_pct', row='time', col='smoker', kind='bar', data=tips[tips.tips_pct < 1])

# set the plotting style
sns.set(style='darkgrid')

# Set plot label anbd title
plt.show()  
```

```python
sns.catplot(x='tips_pct', y='day', kind='box', data=tips[tips.tips_pct < 0.5])


# set the plotting style
sns.set(style='darkgrid')

# Set plot label anbd title
plt.show()  
```
