*QUESTION 1B*

```python
df = pd.read_csv('PT02_1.csv')
df.columns

labels = ['Male Pilgrims','Female Pilgrims']
male=df['Male Pilgrims'].sum()
female=df['Female Pilgrims'].sum()
colors = ['#bd514a','#4f539c']
plt.figure(figsize=(16,9))
plt.pie([male,female],labels=labels,colors=colors,autopct='%.2f%%')
plt.title('Hajj pilgrims by gender 2005 to 2019')
plt.legend()
plt.show()
```

*QUESTION 1C*

```python
df = pd.read_csv('PT02_1.csv')
print(df.columns)

df = df[(df['Year']>=2018) & (df['Year']<=2019)]

labels = ['Male Pilgrims','Female Pilgrims']
male=df['Male Pilgrims'].sum()
female=df['Female Pilgrims'].sum()
colors = ['#bd514a','#4f539c']
plt.figure(figsize=(16,9))
plt.pie([male,female],labels=labels,colors=colors,autopct='%.0f%%')
plt.title('Hajj pilgrims by gender 2018 to 2019')
plt.legend()
plt.show()
```

*QUESTION 2A*

```python
df = pd.read_csv('PT02_3.csv')

labels = ['Brunei/Muara','Tutong','Belait','Temburong']
plt.figure(figsize=(16,9))
plt.boxplot(df[['Brunei/Muara','Tutong','Belait','Temburong']],labels=labels)
plt.title('Chicken Meat Production')
plt.ylabel('Production')

plt.show()



# Two Insight:
#1. Brunei/Muara has the highest Production
#2. Temburong has the lowest prodcution
```

*QUESTION 2B*

```python
df = pd.read_csv('PT02_3.csv')
df['Total Month'] = df['Brunei/Muara'] + df['Tutong'] + df['Belait']+ df['Temburong']
df.head()


labels = df['Month']
values = df['Total Month']


plt.figure(figsize=(14,6))

MyBar = plt.bar(labels,values)
plt.xticks(labels)
plt.yticks()
plt.title('Chicken meat Production')
plt.xlabel('Months')
plt.ylabel('Production in KG')
plt.show()
```

*Question 3A*

```python
df = pd.read_csv('PT02_3.csv')

plt.figure(figsize=(16,9))
plt.plot (df['Month'] ,df['Brunei/Muara'],label = "Brunei/Muara")
plt.plot (df['Month'] ,df['Tutong'],label = "Tutong")
plt.plot (df['Month'] ,df['Belait'],label = "Belait")
plt.plot (df['Month'] ,df['Temburong'],label = "Temburong")
plt.title('Data Distribution Between Four Districts')
plt.xlabel('Month')
plt.ylabel('Value')
plt.legend()

plt.show()

# VI = Brunei/Muara is Octorber , Tutong is September
# VII = From the graph, We can see that from August to September, Tutong production starting to increasing, while Brunei/Muara and Belait descreasing. 
```
