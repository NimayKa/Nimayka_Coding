```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
gas = pd.read_csv("gas_prices.csv")
plt.title('Gas prices')
plt.plot(gas.Year, gas.USA, label="USA")
plt.plot (gas.Year, gas['South Korea'],label = "S.Korea")
plt.xticks(gas.Year[::3])
plt.legend()
plt.show()
```

```python
plt.title('Gas prices')
plt.plot(gas.Year, gas.USA,'b.-', label="USA")
plt.plot (gas.Year, gas['South Korea'],'r.-',label = "S.Korea")
plt.xticks(gas.Year[::3])
plt.legend()
plt.show()
```

```python
plt.title('Gas Price')

for country in gas:
    if country !='Year':
        plt.plot(gas.Year,gas[country],marker='.',label=[country])
plt.xticks(gas.Year[::3])

plt.xlabel('Year')
plt.ylabel('US Dollar')
plt.legend()
plt.show()
```

```python
countries_choesen = ['Germany','Italy','UK','USA']

for country in gas:
    if country in countries_choesen:
        plt.plot(gas.Year,gas[country],marker='.',label=[country])
plt.xticks(gas.Year[::3])

plt.xlabel('Year')
plt.ylabel('US Dollar')
plt.legend()
plt.show()
```
