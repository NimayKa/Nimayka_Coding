```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Cleaned_data.csv")
df.head(5)
```

```python

Retailer_counts = df['Retailer'].value_counts()

plt.figure(figsize=(16, 9))
plt.pie(Retailer_counts, labels=Retailer_counts.index, autopct='%1.1f%%', startangle=140)
plt.legend(loc='best', labels=[f'{label}: {count}' for label, count in zip(Retailer_counts.index, Retailer_counts)])
plt.title('Retailer', fontsize=15)
plt.axis('equal')
plt.savefig('Retailer_counts.png',dpi=200)
plt.show()
```

```python
top_products= df[['Product','Total Sales']].groupby('Product').sum()

top_products.plot(kind='bar',figsize=(16, 9),title='Total sales per Product')

plt.ylabel('Total Sales', size = 15)
plt.savefig('Total_sales_per_Product.png',dpi=200)
plt.show()
```

```python
top_retailers= df[['Retailer','Units Sold']].groupby('Retailer').sum()
top_retailers

plt.figure(figsize=(16, 9))
plt.pie(top_retailers['Units Sold'], labels=top_retailers.index, autopct='%1.1f%%', startangle=140)
plt.legend(loc='best', labels=[f'{label}: {count}' for label, count in zip(top_retailers.index, top_retailers['Units Sold'])])
plt.title('Units Sold per Retailer', fontsize=15)
plt.axis('equal')

plt.savefig('Units_Sold_per_Retailer.png',dpi=200)
plt.show()
```

```python
top_salesmt= df[['Sales Method','Total Sales']].groupby('Sales Method').sum()

plt.figure(figsize=(16, 9))
plt.barh(top_salesmt.index, top_salesmt['Total Sales'])
plt.xticks(fontsize=10)
plt.title('Top Selling Methods', fontsize=25)
plt.savefig('Top Selling Methods.png',dpi=200)
plt.show()
```

```python
best=df[['City','Operating Profit']].groupby('City').sum()
plt.figure(figsize=(16,9))
plt.bar(best.index,best['Operating Profit'])
plt.xticks(best.index,rotation=90)
plt.title('Operating Profit share by City', size = 15)
plt.savefig('Operating_Profit_share_by_City.png',dpi=200)
plt.show()
```

```python
plt.figure(figsize=(16,9))
Margin_per_Retailer= df[['Retailer','Operating Margin']].groupby('Retailer').mean()
print (Margin_per_Retailer)
plt.bar(Margin_per_Retailer.index,Margin_per_Retailer['Operating Margin'])
plt.xticks(Margin_per_Retailer.index)
plt.savefig('Operating_Margin_per_Retailer.png',dpi=200)
plt.show()
```
